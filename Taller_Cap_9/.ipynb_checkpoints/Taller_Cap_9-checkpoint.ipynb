### 9.4) Use el método gráfico para resolver el sistema siguiente
$4x_{1}-8x_{2}=-24\\x_{1}+6x_{2}=34$

*Respuesta:*  
$x_{1}=4\\x_{2}=5$  

*Comprobación: *  

Primer Ecuación  
$4(4)-8(5)=-24\\16-40=-24$  

Segunda Ecuación  
$4+6(5)=34\\4+30=34$

```python
import math
import numpy as np
import matplotlib.pyplot as ptl
%matplotlib inline

ptl.title("Ejercicio 9.4")    
x=np.arange(2,8,0.1)
y=(2*x)-6
z=(-6*x)+34
ptl.plot(x,y,'-g',label="X1=2X2-6")
ptl.plot(x,z,'-y',label="X1=-6X2+34")
ptl.legend(loc="upper left")
ptl.grid(True)
ptl.show()
```

#### d) Resuelva por medio de la eliminación de incógnitas.

$-1.1x_{1}+10x_{2}=120\\-2x_{1}+17.4x_{2}=174$

Despejando $x_{2}$ tenemos:  

$\begin{array}
&& (-2)1.1x_{1} & + & (-2)10x_{2} &  = & (-2)120 \\
- \\
& (-1.1)-2x_{1} & + & (-1.1)17.4x_{2} & = & (-1.1)174\\
\hline
& 0 & + & -20x_{2} + 19.14x_{2} & = & -240 + 191.4
\end{array}$

$x_{2}=\frac{-240+191.4}{-20+19.14}$  

$x_{2}=56.51$

$\begin{array}
&& (17.4)1.1x_{1} & + & (17.4)10x_{2} &  = & (17.4)120 \\
- \\
& (10)-2x_{1} & + & (10)17.4x_{2} & = & (10)174\\
\hline
& -19.14x_{1} + 20x_{1} & + & 0 & = & 2088-1740
\end{array}$

$x_{1}=\frac{2088-1740}{-19.14+20}$  

$x_{1}=404.65$

#### d) Resuelva por medio de la eliminación de incógnitas.

*Respuesta: *

$0.5x_{1}-x_{2}=-9.5\\
1.02x_{1}-2x_{2}=-18.8$

Despejando $x_{2}$ tenemos:  

$\begin{array}
&& (1.02)0.5x_{1} & - & (1.02)x_{2} &  = & (1.02)-9.5 \\
- \\
& (0.5)1.02x_{1} & - & (0.5)2x_{2} & = & (0.5)-18.8\\
\hline
& 0 & + & -1.02x_{2} + x_{2} & = & -9.69 + 9.4
\end{array}$

$x_{2}=\frac{-9.69+9.4}{-1.02+1}$  

$x_{2}=14.5$  

Despejando $x_{1}$ tenemos:  

$\begin{array}
&& (-2)0.5x_{1} & - & (-2)x_{2} &  = & (-2)-9.5 \\
- \\
& (-1)1.02x_{1} & - & (-1)2x_{2} & = & (-1)-18.8\\
\hline
& -x_{1} + 1.02x_{1} & + & 0 & = & 19 - 18.8
\end{array}$

$x_{1}=\frac{19-18.8}{-1+1.02}$  

$x_{1}=10$

Valores de x:

$x_{1}=10$  
$x_{2}=14.5$
