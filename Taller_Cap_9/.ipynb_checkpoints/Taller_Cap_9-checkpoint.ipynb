### 9.4) Use el método gráfico para resolver el sistema siguiente  
$4x_{1}-8x_{2}=-24$  
$x_{1}+6x_{2}=34$

*Respuesta:*  
$x_{1}=4$  
$x_{2}=5$  

*Comprobación: *  

Primer Ecuación  
$4(4)-8(5)=-24$  
$16-40=-24$  

Segunda Ecuación  
$4+6(5)=34$  
$4+30=34$

```python
import math
import numpy as np
import matplotlib.pyplot as ptl
%matplotlib inline

ptl.title("Ejercicio 9.4")    
x=np.arange(2,8,0.1)
y=(2*x)-6
z=(-6*x)+34
ptl.plot(x,y,'-g',label="X1=2X2-6")
ptl.plot(x,z,'-y',label="X1=-6X2+34")
ptl.legend(loc="upper left")
ptl.grid(True)
ptl.show()
```

$0.5x_{1}-x_{2}=-9.5$

$\begin{array}&&(1.02)0.5x_{1}&-&(1.02)x_{2}&=&(1.02)-9.5\\-\\&(0.5)1.02x_{1}&-&(0.5)2x_{2}&=&(0.5)-18.8\\\hline&0&+&-1.02x_{2}+x_{2}&=&-9.69+9.4\end{array}$

$x_{1}=10$
